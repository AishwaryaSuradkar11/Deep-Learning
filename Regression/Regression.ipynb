## Create own dataset with scikit learn

```python
from sklearn.datasets import make_regression
X,Y = make_regression(n_features=30, n_samples=5000, noise=10, random_state=42)
```

```python
X
```

```python
Y
```

```python
X.shape
```

```python
Y.shape
```

```python
cols = [f'x{i}' for i in range(30)]
```

```python
cols
```

```python
import pandas as pd
X = pd.DataFrame(X)
X.columns = cols
```

```python
X
```

```python
Y = pd.DataFrame(Y)
Y.columns = ['Y']
Y
```

## Create a preprocessing pipeline for X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute',SimpleImputer(strategy='median')),
                           ('scaler',StandardScaler())])
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre
```

```python
cols = num_pipe.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre,columns=cols)
X_pre
```

## Train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre,Y,test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

## Create a Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialize the sequential layer
model = Sequential()

# Create a Hidden Layer 1
model.add(Dense(units=32, activation='relu', input_dim = xtrain.shape[1]))

# Create a Hidden Layer 2
model.add(Dense(units=16, activation='relu'))

# Create a Hidden Layer 3
model.add(Dense(units=8, activation='relu'))

# Create Output Layer
model.add(Dense(units=1, activation='linear'))
```

## Compile the model

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

## Fit the model

```python
hist = model.fit(xtrain,ytrain,validation_split=0.2, epochs=100)
```

```python
hist.history['mae']
```

```python
hist.history['val_loss']
```

```python
hist.history['mae']
```

```python
hist.history['val_mae']
```

## Plot the learning curve

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(hist.history['loss'],label = 'Training MSE')
plt.plot(hist.history['val_loss'],label = 'Validation MSE')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.title('Learning Curve')
plt.show()
```

## Predict the values of train & test data

```python
ypred_tr = model.predict(xtrain)
ypred_ts = model.predict(xtest)
```

```python
from sklearn.metrics import r2_score
r2_score(ytrain,ypred_tr)
```
